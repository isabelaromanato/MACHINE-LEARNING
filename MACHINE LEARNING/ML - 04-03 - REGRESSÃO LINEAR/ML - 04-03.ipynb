```python
# IMPORTAR BIBLIOTECAS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm
```

```python
# LER ARQUIVO .CSV

df = pd.read_csv('/content/water_consumption.csv', sep='\t')
df['delta'] = df ['final'] - df['initial']
df
```

```python
# ELIMINAR AS LINHAS COM VALORES AUSENTES
df = df.dropna(how='any', axis=0)
```

```python
# exclua as linhas:  51, 187, 325 do df

df = df.drop([51, 187, 325])
```

```python
# ESCOLHA DAS VARIÁVIES
x = df[['delta']].to_numpy()
y = df['price'].to_numpy() # VARIÁVEL DE INTERESSE
```

```python
# MODEL RLM
model = LinearRegression(n_jobs=None, fit_intercept=True, copy_X=True)
```

```python
print(y[:5])
```

```python
# arrumando valores com problemas (str e float)

df = pd.DataFrame(y)
df[0] = df[0].str.replace('.', '', regex=False)
df[0] = df[0].str.replace(',', '.', regex=False)
y = df[0].astype(float)
```

```python
model.fit(x, y)
```

```python
print(model.score(x, y))
print(model.intercept_)
print(model.coef_)
```

```python
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
model.summary()
```
