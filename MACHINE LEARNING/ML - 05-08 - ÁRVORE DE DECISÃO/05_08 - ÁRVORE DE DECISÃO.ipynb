Vamos usar a base de estudantes.

A base possui 3 variáveis: **Passed**, que indica se um estudante passou de ano, **Study Hours**, que indica quantas horas de estudo o dado aluno teve de dedicação e por fim, **Read Literature**, que indica se o estudante leu a literatura sugerida.

**Passed** é a variável de interesse. **Study Hours** e Read Literature são as variáveis dependentes.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier

df = pd.read_csv('/content/Base Arvore de Decisão - estudantes.csv')
df
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['Read Literature','Study Hours']], df[['Passed']], test_size = 0.5)
```

```python
from sklearn import metrics

regressor = DecisionTreeClassifier()
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)

# Accurary: é o número de previsões corretas dividido pelo total de previsões
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(regressor, out_file=dot_data,
                filled=True, rounded=True,
                special_characters=True, feature_names = ['Read Literature','Study Hours'],class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('image.png')
Image(graph.create_png())
```

```python
from sklearn import metrics

regressor = DecisionTreeClassifier(criterion='log_loss', min_samples_leaf=1, max_depth=1)
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)

# Accurary: é o número de previsões corretas dividido pelo total de previsões
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(regressor, out_file=dot_data,
                filled=True, rounded=True,
                special_characters=True, feature_names = ['Read Literature','Study Hours'],class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('image.png')
Image(graph.create_png())
```
