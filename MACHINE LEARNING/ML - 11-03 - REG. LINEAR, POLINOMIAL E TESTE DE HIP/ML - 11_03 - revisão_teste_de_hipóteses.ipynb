```python
import numpy as np
from scipy import stats

# TESTE DE HIPÓTESE
# ESTUDOS QUANTITATIVOS
# DETERMINAR SE A MÉDIA DE ALTURA DOS ALUNOS DA FIAP É IGUAL A MÉDIA DE ALTURA DA POPULAÇÃO
# AMOSTRA DOS DADOS POPULAÇÃO
# AMOSTRA DOS DADOS DOS ALUNOS DA FIAP
# DISTRIBUIÇÃO NORMAL DOS ERROS
#
# TESTES PARAMÉTRICOS -> TESTES QUE NECESSITAM DE DISTRIBUIÇÃO NORMAL
# TESTES NÃO PARAMÉTRICOS -> NÃO NECESSITA DE DISTRIBUIÇÃO NORMAL
#
# TESTE DE HIPÓTESE É UTILIZADO PARA DETERMINAR IGUALDADE OU DIFERENÇA ***SIGNIFICATIVAMENTE ESTATÍSTICA***
# 1,70m => MÉDIA DE ALTURA DOS ALUNOS DA FIAP
# 1,65m => MEDIA DE ALTURA DA POPULAÇÃO BRASILEIRA
# MATEMATICAMENTE E ESTATÍSTICA É DIFERENTE

#    MÉDIA DA FIAP == MÉDIA DO BRASIL
#    MÉDIA DA FIAP != MÉDIA DO BRASIL

# H0: MÉDIA FIAP == MÉDIA BRASIL
# H1: MÉDIA FIAP != MÉDIA BRASIL
# ALPHA OU SIGNIFICÂNCIA ESTATÍSTICA

# ALPHA: 1%, 5% OU 10%
# MINHA CERTEZA É DE 99% OU 95% OU 90%

# A MÉDIA DA FIAP É IGUAL A MÉDIA DO BRASIL COM 1% DE SIGNIFICÂNCIA ESTATÍSTICA
# A MÉDIA DA FIAP É DIFERENTE DA MÉDIA DO BRASIL EM QUALQUER SIGNIFICÂNCIA ESTATÍSTICA

# UM EMPRESÁRIO, QUE TEM FÁBRICAS DE BOLAS, QUER VERIFICAR SE O DIÂMETRO MÉDIA DAS BOLAS
# PRODUZIDAS PELA FÁBRICA 1 É IGUAL AO DIÂMETRO MÉDIO DAS BOLAS PRODUZIDAS PELA FÁBRICA 2

# DIÂMETRO É IGUAL NAS DUAS FÁBRICAS
# DIÂMETRO É DIFERENTE NAS DUAS FÁBRICAS

# MODELO DO TESTE DE HIPÓTESES
# H0: DIÂMETRO DAS DUAS FÁBRICAS É IGUAL (ALTERNATIVA MENOS DANOSA)
# H1: DIÂMETRO DAS DUAS FÁBRICAS É DIFERENTE (ALTERNATIVA MAIS DANOSA)

# TESTE DE HIPÓTESE USANDO TESTE-T (TESTE PARAMÉTRICO, QUE UTILIZA DISTRIBUIÇÃO NORMAL)
# TESTE DE HIPÓTESE: TEORIA
# TESTE-T TESTE ESTATÍSTICO QUE É UTILIZADO POR UM TESTE DE HIPÓTESE
# TESTE-F
# TESTE QUI-QUADRADO
# ANOVA

# 1ª CONDIÇÃO PARA TESTE-T: NORMALIDADE DOS DADOS
# 2ª CONDIÇÃO PARA TESTE-T: TEM QUE TER HOMOCEDASTICIDADE

amostra_fabrica_1 = [ 9.51182903, 7.19439452, 10.42818857, 8.48623977, 9.95392617, 9.08584036, 7.88828057, 9.13283083, 9.18172261, 7.04365586, 8.57382351, 8.81922503, 6.65734872, 9.14940375, 8.73708844, 9.24066128, 9.99249732, 8.03161223, 10.90146987, 9.35144555, 8.96127198]
amostra_fabrica_2 = [11.10428279, 8.64756835, 10.3397984 , 8.89038564, 9.48686728, 10.71393526, 8.54601478, 10.58205223, 8.90051481, 9.54782322, 9.48101232, 9.22949207, 10.29398161, 11.69433187, 10.25764347, 9.06374192, 10.21169767, 9.51190852, 8.83224818, 9.57369906, 10.61416977, 12.06320768, 9.50302557, 8.82285133, 9.15629123]
alpha = 0.05

# TESTE DE SHAPIRO
# H0: DIZ QUE OS DADOS NÃO SÃO NORMAIS
# H1: DIZ QUE OS DADOS SÃO NORMAIS

estatistica_t, p_valor_1 = stats.shapiro(amostra_fabrica_1)
estatistica_t, p_valor_2 = stats.shapiro(amostra_fabrica_2)

if (p_valor_1 < alpha && p_valor_2 < alpha):
	print("Shapiro: ambas amostras seguem distribuição normal")

  # TESTE DE LEVENE
  # H0: DIZ QUE OS DADOS TEM HETEROCEDASTICIDADE
  # H1: DIZ QUE OS DADOS TEM HOMOCEDASTICIDADE

	estatistica_t, p_valor_1 = stats.levene(amostra_fabrica_1, amostra_fabrica_2)

	if (p_valor_1 < alpha):
		print("Levene: tem HOMOCEDASTICIDADE")

    estatistica_t, p_valor = stats.ttest_ind(amostra_fabrica_1, amostra_fabrica_2)

		# modelar o teste de hipóteses
		# H0: média1 == média2
		# H1: média1 != média2
		print("P-valor: ", p_valor)
		print("Estatística-t: ", estatistica_t)

    # P-VALOR: É CERTEZA DO MEU CÁLCULO; QUANTO MENOR O P-VALOR, MAIS CERTO ESTOU DO RESULTADO
    # P-VALOR É COMPARÁVEL AO ALPHA
    # P-VALOR É MENOR DO QUE O ALPHA ENTÃO H0 É REJEITADO E H1 ACEITO
    # P-VALOR É MAIOR OU IGUAL AO ALPHA ENTÃO H0 É ACEITO E H1 REJEITADO

    # MODELO DO TESTE DE HIPÓTESES
    # H0: DIÂMETRO DAS DUAS FÁBRICAS É IGUAL (ALTERNATIVA MENOS DANOSA)
    # H1: DIÂMETRO DAS DUAS FÁBRICAS É DIFERENTE (ALTERNATIVA MAIS DANOSA)
    if p_valor < alpha:
      print("H0 rejeitado, logo, diâmetro é diferente nas duas fábricas")
    if p_valor >= alpha:
      print("H0 aceito, logo, diâmetro é igual nas duas fábricas")
	else:
		print("Levene: não tem HOMOCEDASTICIDADE")
else:
	print("Shapiro: as amostras não seguem distribuição normal")



```
