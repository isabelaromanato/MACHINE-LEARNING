Vamos usar a base de estudantes.

A base possui 3 variáveis: **Passed**, que indica se um estudante passou de ano, **Study Hours**, que indica quantas horas de estudo o dado aluno teve de dedicação e por fim, **Read Literature**, que indica se o estudante leu a literatura sugerida.

**Passed** é a variável de interesse. **Study Hours** e Read Literature são as variáveis dependentes.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

df = pd.read_csv('/content/drive/MyDrive/datasets/Base Arvore de Decisão - estudantes.csv')
df.head(2)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['Read Literature','Study Hours']],
                                                    df[['Passed']], test_size = 0.5)
```

```python
from sklearn import metrics

clf = RandomForestClassifier(oob_score=True, n_estimators=10, random_state=42)
clf.fit(x_train, y_train.values.ravel())
y_pred = clf.predict(x_test)

# Accurary: é o número de previsões corretas dividido pelo total de previsões
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("OOB_score:", clf.oob_score_)
```
