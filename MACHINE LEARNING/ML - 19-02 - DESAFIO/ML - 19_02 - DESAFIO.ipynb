```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm
```

```python
dados = pd.read_csv('base.csv',
            encoding = 'latin1',
            sep = ';',
            decimal = ',')
```

```python
# X: as minhas variáveis explicativas;
x = dados[['idade','amigos',	'esportes',	'balada',	'teatro',	'cinema',	'musicais','filmes_casa']].to_numpy()
```

```python
# Y: é um array unidimensional, pois só há 1 variável sendo prevista
y = dados['gastos'].to_numpy()
```

```python
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.summary())
```
