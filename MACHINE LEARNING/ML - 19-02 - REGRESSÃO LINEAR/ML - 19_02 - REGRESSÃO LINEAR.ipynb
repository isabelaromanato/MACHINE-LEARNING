### Forma de uma Regressão Linear (equação de 1° grau: )

Y = b0 + b1 * X

```python
# Declarando o model:
# n jons indica a qtd de processos paralelos que vão calcular a  reglin
# fit_intercept = False, indica que não há constante na equação criada; uso é raro!
                                      # EU QUERO TER UMA CONSTANTE
model = LinearRegression(n_jobs=None, fit_intercept=True)
```

```python
# predizer o salário através dos anos de experiência

# X: São as minhas variáveis explicativas
x = df['anos_exp'].to_numpy().reshape(-1,1)

# Y: é um array unidimensional, pois só há 1 variável
y = df['salario'].to_numpy() # sem reshape pois preciso de 1 previsã
```

```python
# faz a modelagem da regressão linear, calculando o intercepto e o coeficiente angular
model.fit(x, y)
```

```python
# score calcula o r-quadrado
print(model.score(x, y))
```
