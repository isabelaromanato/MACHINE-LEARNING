### Forma de uma Regresso Linear (equao de 1 grau: )

Y = b0 + b1 * X

```python
# Declarando o model:
# n jons indica a qtd de processos paralelos que vo calcular a  reglin
# fit_intercept = False, indica que no h constante na equao criada; uso  raro!
                                      # EU QUERO TER UMA CONSTANTE
model = LinearRegression(n_jobs=None, fit_intercept=True)
```

```python
# predizer o salrio atravs dos anos de experincia

# X: So as minhas variveis explicativas
x = df['anos_exp'].to_numpy().reshape(-1,1)

# Y:  um array unidimensional, pois s h 1 varivel
y = df['salario'].to_numpy() # sem reshape pois preciso de 1 previs
```

```python
# faz a modelagem da regresso linear, calculando o intercepto e o coeficiente angular
model.fit(x, y)
```

```python
# score calcula o r-quadrado
print(model.score(x, y))
```
