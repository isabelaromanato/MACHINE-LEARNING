```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv('salarios2.csv', sep='\t')
```

```python
# Forma de uma Regressão Linear (equação de 1º grau, lembra? :)
# Y = b0 + b1 * X
```

```python
# X: são as minhas variáveis explicativas; tem que
# ser produzido como array bidimensional
# para acomodar múltiplas variáveis por observação.
x = df[['anos_exp','anos_estudo']].to_numpy()
```

```python
# Y: é um array unidimensional, pois só há 1 variável
# sendo prevista
y = df['salario'].to_numpy()
```

```python
# Declarando o modelo;
# n_jobs indica a quantidade de processos paralelos que vão calcular a reglin
# fit_intercept = False, indica que não há constante na equação criada; uso é raro!
model = LinearRegression(n_jobs=None, fit_intercept=True)
```

```python
# faz a modelagem da regressão linear, calculando o
# intecepto e o coeficiente angular
model.fit(x, y)
```

```python
# score calcula o r-quadrado
print(model.score(x, y))
# RLS: 0.9854136517127601
# RLM: 0.9868462044575527
```
