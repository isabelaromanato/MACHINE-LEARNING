```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
import scipy.stats as stats
```

```python
file = '/content/drive/MyDrive/datasets/Colesterol - Colesterol.csv'
df = pd.read_csv(file)
df
```

```python
df.isnull().sum()
```

```python
df1 = df
df_filtered = df1.query('HDL.notnull() and Total.notnull()')
```

```python
x = df_filtered[['HDL', 'Total']]
y = df_filtered['Idade'].to_numpy()
model = LinearRegression()
model.fit(x, y)
score_original = model.score(x, y)
print(score_original)
```

```python
# Vou começar a reconstruir HDL
# Vou utilizar como variável regressora Total
#IDADE NÃO PODE!!!

df_filtered = df1[ (df1['Total'].notnull()) & (df1['HDL'].notnull()) ]
x = df_filtered[['Total']].to_numpy()
y = df_filtered['HDL'].to_numpy()
model_hdl = LinearRegression()
model_hdl.fit(x, y)
score = model_hdl.score(x, y)
score
```

```python
# minha função criada para completar o HDL de cada linha
def completar_hdl(linha):
  if pd.isna(linha['HDL']) and not pd.isna(linha['Total']):
    campos = linha[['Total']]
    return round(model_hdl.predict([campos])[0])
  return linha['HDL']

# apply: executar a função chamada 1x para cada linha
df1['hdl_calculado'] = df1.apply(completar_hdl, axis=1)
df1
```

```python
df_filtered = df1[ (df1['Total'].notnull()) & (df1['hdl_calculado'].notnull()) ]
x = df_filtered[['hdl_calculado']].to_numpy()
y = df_filtered['Total'].to_numpy()
model_total = LinearRegression()
model_total.fit(x, y)
score = model_total.score(x, y)
score
```

```python
# minha função criada para completar o Total
def completar_total(linha):
  if pd.isna(linha['Total']) and not pd.isna(linha['hdl_calculado']):
    campos = linha[['hdl_calculado']]
    return round(model_total.predict([campos])[0])
  return linha['Total']

# apply: executar a função chamada 1x para cada linha
df1['total_calculado'] = df1.apply(completar_total, axis=1)
df1
```

```python
# Recalculo a minha regressão original utilizando as variáveis novas
x = df1[['hdl_calculado', 'total_calculado']]
y = df1['Idade'].to_numpy()
model = LinearRegression()
model.fit(x, y)
score_reconstruido = model.score(x, y)
print(score_reconstruido)
```

```python
print("O score original resultou em:", score_original, "e já o score reconstruído em:", score_reconstruido)
```

Informações importantes da aula:

- Se a reta se aproxima muito do dado, o R-Quadrado tende a ser maior.
- Quanto maior a amostra, mais chances de outliers e possibilidade de puxar o R- Quadrado para baixo.
- Quanto mais pontos eu tenho, maior a confiabilidade terei naquela Regressão.

---

**Ao observar o dataset e já arrancar os nulos**

- Não é viável dropar os nulos de todas as colunas, para não perder os dados.
- Observar colunas importantes para os cálculos e regressões e filtrar excluindo os nulos dessas colunas.
- Ex: excluir todos os nulos do dataset e tinha valores nulos apenas na coluna "NOME" (não é possível utilizar na regressão). Perdemos dados atoa.

---

**Após filtrar perdi uma porcentagem alta da base de dados**

- Ex: perdemos 40% da base de dados após filtros.
- Em uma base pequena isso é crítico.
- Embora o R- Quadrado tende a ser maior, a confiabilidade é baixa.
