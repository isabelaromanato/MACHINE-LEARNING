Utilizar a base colesterol.csv para:

1. Calcular regressão para determinar a idade, filtrando nulos.
2. Reconstruir missings utilizando regressões.
3. Recalcular a regressão original com os missings reconstruídos
4. Comparar o r2 da regressão com dados filtrados e da regressão com os missings reconstruídos.

```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
import scipy.stats as stats

file = '/content/drive/MyDrive/datasets/Colesterol - Colesterol.csv'
df = pd.read_csv(file)
df
```

```python
df_filtered = df[(df['HDL'].notnull()) & (df['Total'].notnull())]
df_filtered = df.query('Total.notnull() and HDL.notnull()')
df_filtered = df.dropna()
df_filtered

x = df_filtered[['Total','HDL']]
y = df_filtered['Idade'].to_numpy()

model = LinearRegression()
model.fit(x, y)
score = model.score(x, y)
score
```

```python
# Não posso pegar a Idade de jeito maneira.
# Isso porque a variável de interesse (Idade) tem que ser independente, ou seja,
# não posso utilizá-la para regredir os regressores!!!
df_filtered = df[(df['HDL'].notnull()) & (df['Total'].notnull())]
x = df_filtered[['HDL']].to_numpy()
y = df_filtered['Total'].to_numpy()
model_total = LinearRegression()
model_total.fit(x, y)
score_total = model_total.score(x, y)
score_total
```

```python
def calcular_total(linha):
  if pd.isna(linha['Total']) and not pd.isna(linha['HDL']):
    campos = linha[['HDL']]
    return model_total.predict([campos])[0]
  return linha['Total']

df['Total Calculado'] = df.apply(calcular_total, axis=1)
df.head(5)
```

```python
# Não posso pegar a Idade de jeito maneira.
# Isso porque a variável de interesse (Idade) tem que ser independente, ou seja,
# não posso utilizá-la para regredir os regressores!!!
df_filtered = df[(df['HDL'].notnull()) & (df['Total'].notnull())]
x = df_filtered[['Total']].to_numpy() # variável regressora!!!
y = df_filtered['HDL'].to_numpy() # variável de interesse!!!
model_hdl = LinearRegression()
model_hdl.fit(x, y)
score_hdl = model_hdl.score(x, y)
score_hdl
```

```python
def calcular_hdl(linha):
  if pd.isna(linha['HDL']) and not pd.isna(linha['Total']):
    campos = linha[['Total']]
    return model_hdl.predict([campos])[0]
  return linha['HDL']

df['HDL Calculado'] = df.apply(calcular_hdl, axis=1)
df
```

```python
df_filtered = df[(df['HDL Calculado'].notnull()) & (df['Total Calculado'].notnull())]

x = df_filtered[['Total Calculado','HDL Calculado']]
y = df_filtered['Idade'].to_numpy()

model = LinearRegression()
model.fit(x, y)
score = model.score(x, y)
score

# 0.15171980214127057 minha chance de overfitting é maior! Pq? Porque tenho menos observações
# 0.14523685084391502
```
